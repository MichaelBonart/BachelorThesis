Analyze the dichotomy of the COAD dataset. (maybe to split it up)

```python
import mhn
import numpy as np
import pandas as pd
import os
import scipy
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
filename="G16_CH_COAD_Events.csv"
file_path=os.path.join( os.getcwd(), '..','data', filename)
raw_data_input = pd.read_csv(file_path)
cols=raw_data_input.columns.str.endswith(')')
del_col_mask=raw_data_input.columns.str.endswith('(Del)')
amp_col_mask=raw_data_input.columns.str.endswith('(Amp)')
mut_col_mask=raw_data_input.columns.str.endswith('(M)') #filter out APC (M) and TP53 (M) ??


copy_n_alt=raw_data_input.loc[:,del_col_mask].sum(axis=1)+raw_data_input.loc[:,amp_col_mask].sum(axis=1)+raw_data_input['APC (M)']+raw_data_input['TP53 (M)']
plt.hist(copy_n_alt,bins=8)
plt.show()
mutations= raw_data_input.loc[:,mut_col_mask].sum(axis=1)-raw_data_input['APC (M)']-raw_data_input['TP53 (M)']
plt.hist(mutations)
plt.show()
plt.scatter(mutations, copy_n_alt, alpha=0.1)
```
